```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
iris=load_iris()
dir(iris)
```

```python
iris.data
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df=df.drop(df.iloc[:,0:2],axis=1)
```

```python
df
```

```python
df
```

```python
df.drop(['flower'],axis=1,inplace=True)
df
```

```python
km=KMeans(n_clusters=3)
km
```

```python
yp=km.fit_predict(df)
```

```python
yp
```

```python
df['cluster']=yp
```

```python
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
plt.scatter(df1.iloc[:,0],df1.iloc[:,1],color='red')
plt.scatter(df2.iloc[:,0],df2.iloc[:,1],color='green')
plt.scatter(df3.iloc[:,0],df3.iloc[:,1],color='blue')
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
```

```python
sse=[]
k_rng=range(1,10)
for i in k_rng:
    km=KMeans(n_clusters=i)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
```
