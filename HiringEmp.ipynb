```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.drop('Unnamed: 2',axis=1,inplace=True,errors='ignore')
df
```

```python
df.drop('Unnamed: 4',axis=1,inplace=True,errors='ignore')
df
```

```python
df.dropna(how="all",inplace=True)
```

```python
df
```

```python
df.drop(8,inplace=True,errors='ignore')
```

```python
df.drop(15,inplace=True,errors='ignore')
df
```

```python
print(df['test_score(out of 10)'])
```

```python
df['test_score(out of 10)']=pd.to_numeric(df['test_score(out of 10)'])
```

```python
df['test_score(out of 10)'].median()
```

```python
df.fillna({'test_score(out of 10)':df['test_score(out of 10)'].median()},inplace=True)
```

```python
df
```

```python
df.fillna({'experience':'zero'},inplace=True)
df
```

```python
from word2number import w2n
df['experience']=df['experience'].astype(str)
df['experience']=df['experience'].apply(w2n.word_to_num)
df
```

```python
Reg=linear_model.LinearRegression()
Reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```
